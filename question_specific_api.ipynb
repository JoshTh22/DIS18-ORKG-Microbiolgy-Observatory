```python
import pandas as pd
from Bio import Entrez, Medline
```

```python
Entrez.email = "joshua.thos@smail.th-koeln.de"
```

```python
query = input("Enter a query: ")
```

```python
pubmed_idlist = []
handle = Entrez.esearch(db="pubmed", term = query, retmax = "10")
record = Entrez.read(handle)
pubmed_idlist.append(record["IdList"])
flat_pubmed_idlist = [item for sublist in pubmed_idlist for item in sublist]
```

```python
flat_pubmed_idlist
```

```python
handle = Entrez.efetch(db = "pubmed", id = flat_pubmed_idlist, rettype = "medline", retmode = "json")
records = Medline.parse(handle)

data_list =[]

for record in records:
    data_list.append(record)
```

```python
data_list
```

```python
df = pd.DataFrame(data_list)
```

```python
df = df[["TI", "AB", "AID"]]
```
