```python
#import all modules that are required

import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
#add the links of the websites we want to parse

url1 = "https://flexikon.doccheck.com/de/Klinisch_relevante_Bakterien"
url2 = "https://en.wikipedia.org/wiki/List_of_clinically_important_bacteria"
```

```python
#open the page and store it in a variable

page1 = urlopen(url1)
page2 = urlopen(url2)
```

```python
#parse the first page and extrace all informations we want to receive

#creating an empty list to store data in it
liste = []

#parse the pages and store the html text in a variable
html_parser1 = bs(page1, 'html.parser')
html_parser2 = bs(page2, 'html.parser')

#search for the specific information we want
content1 = html_parser1.find('div', {"class": "mw-content-ltr"})
content2 = html_parser2.find('div', {"class": "mw-parser-output"})

names1 = ''
names2 = ''

#scrape the content and append it to the list
for i in content1.findAll('li'):
    names1 = i.text.split('\n\n')
    liste.append(names1)
    
for i in content2.findAll('i'):
    names2 = i.text.split('\n')
    liste.append(names2)
    
flat_list = [item for sublist in liste for item in sublist]

flat_list2 = []

for i in flat_list:
    if i != '\n':
        flat_list2.append(i.strip('\n'))
        
#print(flat_list2)        
```

```python
#save the names of the bacteria in a dataframe, drop all duplicates and store the data in a csv file

#create datafrae with the scaped data
df = pd.DataFrame(flat_list2, columns = ['bacteria names'])

#drop all duplactes
df.drop_duplicates().to_csv('bacteria_names_new.csv', index = False)

print(liste)
#save the data in a csv file
#with open('bacteria_names_new.csv', 'w') as file:
#    file.write(liste)
```
